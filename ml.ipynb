```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df1 = pd.read_csv('https://girlsgoitpublic.z6.web.core.windows.net/movies_train.csv')
movies = df1.copy()
movies2 = df1.copy()
movies
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna(subset=['country'])
movies.isnull().sum() 
```

```python
movies = movies.dropna(subset=['language'])
movies.isnull().sum() 
```

```python
for index, row in movies.iterrows():
  if movies.loc[index, 'country'].find(', ') > -1:
    movies.loc[index, 'country'] = 'Multiple countries'
```

```python
movies.head()
```

```python
movies = movies.sample(frac=1)
```

```python
from sklearn.model_selection import train_test_split

X=movies[['duration']] 
y=movies['popularity']  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.feature_extraction.text import CountVectorizer
clf=RandomForestClassifier(n_estimators=50, max_depth=4)

clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
# Acuratețea modelului
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
 from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators': [50, 60, 70, 80, 90, 100],
              'max_depth': [2, 3, 4, 5, 6, 7],
              'criterion': ['gini', 'entropy'],
              }
```

```python
rf = RandomForestClassifier()
```

```python
gscv = GridSearchCV(rf, parameters)
```

```python
gscv.fit(X_train,y_train)
```

```python
gscv.best_score_
```

```python
gscv.best_estimator_
```

```python
movies_ml = movies[['movie_id', 'genre', 'country', 'language', 'director', 'production_company', 'popularity']]
movies_ml
```

```python
movies_ml.isnull().sum()
```

```python
movies_ml['country'].fillna('Unknown', inplace = True)
movies_ml['language'].fillna('Unknown', inplace = True)
movies_ml['production_company'].fillna('Unknown', inplace = True)
movies_ml['director'].fillna('Unknown', inplace = True)
movies_ml.isnull().sum()
```

```python
movies_ml
```

```python
X = movies_ml.select_dtypes(include=[object])
X
```

```python
from sklearn import preprocessing
X.columns
```

```python
le = preprocessing.LabelEncoder()

X2 = X.apply(le.fit_transform)
X2
```

```python
from sklearn.model_selection import train_test_split

X=X2[['genre', 'country', 'director', 'language', 'production_company']] 
y=X2['popularity']  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.feature_extraction.text import CountVectorizer
clf=RandomForestClassifier(n_estimators=90, max_depth=7)

clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
# Acuratețea modelului
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
 from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators': [50, 60, 70, 80, 90, 100],
              'max_depth': [2, 3, 4, 5, 6, 7],
              'criterion': ['gini', 'entropy'],
              }
```

```python
rf = RandomForestClassifier()
```

```python
gscv = GridSearchCV(rf, parameters)
```

```python
gscv.fit(X_train,y_train)
```

```python
gscv.best_score_
```

```python
gscv.best_estimator_
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_clf = AdaBoostClassifier()
```

```python
clf=ada_clf()

clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
# Acuratețea modelului
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
 from sklearn.model_selection import GridSearchCV

 parameters = {'n_estimators': [50, 60, 70, 80, 90, 100],
              'learning_rate': [2, 3, 4, 5, 6, 7],
              'random_state': [2,3,4,5,6],
               'algorithm': ['SAMME', 'SAMME.R']
              }

gscv = GridSearchCV(ada_clf, parameters)
gscv.fit(X_train,y_train)
gscv.best_score_
```
